```python
over = train.loc[train.multi_label=='over']
under = train.loc[train.multi_label=='under']
hit = train.loc[train.multi_label=='hit']
falsepos = train.loc[train.multi_label=='falsepos']

# upsample under,hit,falsepos
upsample_under=resample(under,random_state=3,n_samples=len(over),replace=True)
upsample_hit=resample(hit,random_state=3,n_samples=len(over),replace=True)
upsample_falsepos=resample(falsepos,random_state=3,n_samples=len(over),replace=True)

train_resample = pd.concat([upsample_under,upsample_hit,upsample_falsepos,over])

# reshuffle
train_resample = train_resample.sample(frac=1,random_state=1)

X_train_resample = train_resample.drop(columns=['multi_label'])
y_train_resample = train_resample['multi_label']
```

```python
# baseline model using 5 fold x validation
# see if results are about the same for different samples of dataset

baseline = BaggingClassifier()

baseline_cv = cross_validate(baseline, X_train_resample, y_train_resample, cv = kfold, scoring="roc_auc")

print(baseline_cv['test_score'])
print('Mean ROC AUC: %.3f' % baseline_cv['test_score'].mean())
```

### Look at other model eval metrics

correlation matrix

```python
def plot_cm(labels, predictions):
    cm = confusion_matrix(labels, predictions)
    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title('Confusion matrix')
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')

    print('Legitimate Transactions Detected (True Negatives): ', cm[0][0])
    print('Legitimate Transactions Incorrectly Detected (False Positives): ', cm[0][1])
    print('Fraudulent Transactions Missed (False Negatives): ', cm[1][0])
    print('Fraudulent Transactions Detected (True Positives): ', cm[1][1])
    print('Total Fraudulent Transactions: ', np.sum(cm[1]))
```
