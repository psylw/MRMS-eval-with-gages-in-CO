```python

for name, clf in zip(names, classifiers):
    clf = clf

    sampler = RandomOverSampler(random_state=42)
    pipeline = Pipeline([('sampler', sampler),('gbt', clf)])
    pipeline.fit(X_train, y_train)    
    
    # predict probabilities
    yhat = clf.predict_proba(X_test)
    # keep probabilities for the positive outcome only
    probs = yhat[:, 1]
    # define thresholds
    thresholds = np.arange(0, 1, 0.001)
    # evaluate each threshold, use cross fold instead of test data
    scores = [roc_auc_score(y_test, to_labels(probs, t)) for t in thresholds]

    # get best threshold
    ix = np.argmax(scores)
    threshold = thresholds[ix]
    
    print(name+'f1: '+str(f1_score(y_test, to_labels(probs, threshold))))
    print(name+'accuracy: '+str(accuracy_score(y_test, to_labels(probs, threshold))))
    print(name+'ovr_roc_auc: '+str(roc_auc_score(y_test, to_labels(probs, threshold))))
    print(name+'precision: '+str(precision_score(y_test, to_labels(probs, threshold))))
    print(name+'recall: '+str(recall_score(y_test, to_labels(probs, threshold))))
```

### Feature importance

```python
top10_mean = []
for name, clf in zip(names, classifiers):

    clf=clf
    output = cross_validate(clf, data,labels, cv=cv, scoring = 'accuracy', return_estimator =True)
    fi = []
    for idx,estimator in enumerate(output['estimator']):
        feature_importances = pd.DataFrame(estimator.feature_importances_,
                                           index = df.drop(columns='label').columns,
                                            columns=['importance']).sort_values('importance', ascending=False)
        fi.append(feature_importances)
    
    fi_mean = pd.concat(fi,axis=1).mean(axis=1).sort_values(ascending=False)
    top10_mean.append({name:fi_mean[0:10]})
    fig, ax = plt.subplots(figsize=(8,5))
    fi_mean.plot.bar(ax=ax)
    ax.set_title(name)

    fig.tight_layout()
    plt.show()
```

```python
# look at distribution of probabilities for each classifier
predict_proba = []
for name, clf in zip(names, classifiers):

    clf=clf
    pipe = Pipeline([
    ("model", Thresholder(clf, threshold=0.7))
    ])
    
    pipe.fit(data,labels)
    predict_proba.append(pipe.predict_proba(predict_s)[:, 1])
```
